```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Tweets.csv')  # adjust path accordingly

sentiment_counts = df['airline_sentiment'].value_counts()

plt.bar(sentiment_counts.index, sentiment_counts.values, color=['green', 'red', 'gray'])
plt.title('Tweet Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()
```

```python
df['tweet_created'] = pd.to_datetime(df['tweet_created'])
tweets_per_day = df.groupby(df['tweet_created'].dt.date).size()

plt.plot(tweets_per_day.index, tweets_per_day.values)
plt.title('Number of Tweets per Day')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.show()
```

```python
airline_counts = df['airline'].value_counts()

plt.bar(airline_counts.index, airline_counts.values, color='skyblue')
plt.title('Number of Tweets per Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.show()
```

```python
sentiment_airline = pd.crosstab(df['airline'], df['airline_sentiment'])

sentiment_airline.plot(kind='bar', stacked=True, color=['green', 'red', 'gray'])
plt.title('Sentiment Breakdown by Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()
```
